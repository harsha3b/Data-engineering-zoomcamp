```python
zone_data='taxi_zone_lookup.csv'
```

```python
df2=pd.read_csv(zone_data)
```

```python
df2.info()
```

```python
df2.head()
```

```python
null_columns=df2.columns[df2.isnull().any()]
print(null_columns)
```

```python
rows_with_nulls=df2[df2.isnull().any(axis=1)]
print(rows_with_nulls)
```

```python
# fill the missing zone and service_zone fields with 'Unknown'
df2['Zone'] = df2['Zone'].fillna('Unknown')
df2['service_zone'] = df2['service_zone'].fillna('Unknown')

```

```python
# view the corrected rows
print(df2.loc[[263, 264]])
```

```python
# align your data types before exporting to postgresql to keep indexes efficient
df['PULocationID'] = df['PULocationID'].astype('Int64')
df['DOLocationID'] = df['DOLocationID'].astype('Int64')
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, name='green_taxi_data', con=engine))
```

```python
df.head(n=0).to_sql(name='green_taxi_data', con=engine, if_exists='replace')
```

```python
len(df)
```

```python
df.to_sql(name='green_taxi_data', con=engine, if_exists='append', index=False, chunksize=10000)
```
